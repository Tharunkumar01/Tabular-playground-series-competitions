```python
import pandas as pd
import numpy as np
import xgboost as xgb
import lightgbm as lgbm 
import catboost as catboost
import pycaret.regression as reg
```

```python
train = pd.read_csv('../input/tabular-playground-series-aug-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-aug-2021/test.csv')
sample = pd.read_csv('../input/tabular-playground-series-aug-2021/sample_submission.csv')
```

```python
# train.nunique()
```

```python
automl = reg.setup(data=train.drop('id', 1), target='loss', train_size=0.7, preprocess=True, feature_selection=True, fold_strategy='stratifiedkfold', use_gpu=True, silent=True)
```

```python
models = reg.compare_models(sort='rmse', n_select = 3, budget_time=120)
```

```python
ensemble = reg.blend_models(models, choose_better=True, optimize = 'rmse')
```

```python
stack = reg.stack_models(models, choose_better=True, optimize ='rmse')
```

```python
ensemble_preds = reg.predict_model(ensemble, data = test.drop('id', 1))
stack_preds = reg.predict_model(stack, data = test.drop('id', 1))
```

```python
ensemble_preds.to_csv('ensemble_preds.csv', index=False)
stack_preds.to_csv('stack_preds.csv', index=False)
```
