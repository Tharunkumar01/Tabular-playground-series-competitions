```python
import pandas as pd
import numpy as np
import xgboost as xgb 
import lightgbm as lgm
import catboost as cat
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as mae
from sklearn.model_selection import StratifiedKFold
```

```python
train = pd.read_csv('../input/tabular-playground-series-aug-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-aug-2021/test.csv')
sample = pd.read_csv('../input/tabular-playground-series-aug-2021/sample_submission.csv')
```

```python
X = train.drop(labels=['id','loss'],axis=1,inplace=False).copy()
y = train['loss']
X_test = test.drop(labels=['id'],axis=1,inplace=False).copy()
random_seed = 2001
```

```python
params = {
        'booster': 'dart',
        'rate_drop': 0.1, 
        'eta': 0.1, 
        'deterministic_histogram': False, }
```

```python
params_xgb = params
params_xgb["tree_method"] = "gpu_hist"
params_xgb["predictor"] = 'gpu_predictor'

test_preds_list = []
score_list = []
oof = np.zeros((len(train)))
fold = 1

kf = StratifiedKFold(n_splits=3,shuffle=True,random_state=2001)
splits = list(kf.split(X,y))

for train_idx, val_idx in splits:
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    val_preds_list = []
    
    # fit and run model
    params_xgb['seed'] = 2001

    dtrain = xgb.DMatrix(data=X_train, label=y_train)
    dval = xgb.DMatrix(data=X_val, label=y_val)
    dtest = xgb.DMatrix(data=X_test)

    model = xgb.train(params_xgb, dtrain,\
                       evals=[(dtrain,'train'),(dval,'val')],\
                       verbose_eval=100,
                       early_stopping_rounds=10,
                       num_boost_round=100000)

    val_preds_list.append(model.predict(dval))
    test_preds_list.append(model.predict(dtest))

    oof[val_idx] = np.mean(val_preds_list,axis=0)
    score = mae(y_val, oof[val_idx], squared=False)
    print(f"fold: {fold},rmse: {score}")
    score_list.append(score)
    fold += 1
name = 'xgboost_3seeds_5fold'

cv_logloss = np.mean(score_list)
print(f"{name} ,rmse: {cv_logloss}")

preds= np.mean(test_preds_list,axis=0)

sample['loss'] = preds
sample.to_csv('submission.csv',index=None)
```
